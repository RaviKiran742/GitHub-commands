```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
class Model(nn.Module):
  def __init__(self,input_size,hidden_size1,hidden_size2,num_classes):
    super(Model, self).__init__()
    self.l1=nn.Linear(input_size,hidden_size1)
    self.relu=nn.ReLU()
    self.l2=nn.Linear(hidden_size1,hidden_size2)
    self.r2=nn.ReLU()
    self.l3=nn.Linear(hidden_size2,num_classes)

  def forward(self,x):
    out=self.l1(x)
    out=self.relu(out)
    out=self.l2(out)
    out=self.r2(out)
    out=self.l3(out)
    return out

```

```python
hs1=512
hs2=234
input_size=X.shape[1]
num_classes=1
model=Model(input_size,hs1,hs2,num_classes)
```

```python
model.parameters()
```

```python
optimizer=optim.Adam(model.parameters(),lr=0.001)
criterion=nn.MSELoss()
```

```python
from torch.utils.data import DataLoader,Dataset
```

```python
class MyDataset(Dataset):
    def __init__(self, dataframe,labels):
        self.data = torch.tensor(dataframe.values, dtype=torch.float32)
        self.labels = torch.tensor(labels.values, dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]
```

```python
X_train=MyDataset(X_train,Y_train)
X_test=MyDataset(X_test,Y_test)
train_loader=DataLoader(dataset=X_train,batch_size=64,shuffle=True)
test_loader=DataLoader(dataset=X_test,batch_size=64,shuffle=True)
```

```python
for epoch in range(1000):
  for batch_idx,(data,targets) in enumerate(train_loader):
    scores=model(data).squeeze()
    loss=criterion(scores,targets)


    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if batch_idx % 100 == 0:
            print(f'Epoch [{epoch+1}/{100}], Step [{batch_idx}/{len(train_loader)}], Loss: {loss.item():.4f}')
```
